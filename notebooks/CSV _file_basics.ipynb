8. Use Concat to merge csv files

```python
titanic_train_path = netflix_path = os.path.join(data_path, 'titanic', 'train.csv')
titanic_test_path = netflix_path = os.path.join(data_path, 'titanic', 'test.csv')

titanic_train_df = pd.read_csv(titanic_train_path, index_col='PassengerId')
titanic_test_df = pd.read_csv(titanic_test_path, index_col='PassengerId')

titanic_df = pd.concat((titanic_test_df, titanic_train_df))
```
